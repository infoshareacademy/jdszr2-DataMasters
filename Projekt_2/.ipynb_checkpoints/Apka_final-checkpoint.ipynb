```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
%matplotlib inline
import ipywidgets as widgets
import seaborn as sns
import matplotlib.pyplot as plt
import os 
import datetime as dt
```

```python
file_18 = 'ks-projects-201801.csv' 
file_path = os.path.join(os.path.abspath(""), file_18)

df_2018 = pd.read_csv(file_path)
df_2018 = df_2018.dropna()

del df_2018['ID']
del df_2018['goal']
del df_2018['pledged']
del df_2018['usd pledged']


df_2018['launched'] =  pd.to_datetime(df_2018['launched'])
df_2018['deadline'] =  pd.to_datetime(df_2018['deadline'])
df_2018['project_duration']=(df_2018['deadline'] - df_2018['launched']).map(lambda x: np.nan if pd.isnull(x) else x.days)
```

```python
df_2018 = df_2018[(df_2018.state != "live")]
```

```python
ALL = 'ALL'
def ALL_values(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique
```

```python
# Stworzenie arkuszy 
output = widgets.Output()
plot_output = widgets.Output()
plot_output_1 = widgets.Output()
histogram_days = widgets.Output()
data_output = widgets.Output()



# Dopisanie do widgetów warunku
dropdown_main_category = widgets.Dropdown(options = ALL_values(df_2018.main_category), description = 'main_category')
dropdown_category = widgets.Dropdown(options = ALL_values(df_2018.category), description = 'category')
dropdown_country = widgets.Dropdown(options = ALL_values(df_2018.country), description = 'country')

def common_filtering(main_category, category, country):
    output.clear_output()
    
    if (main_category == ALL) & (category == ALL) & (country == ALL):
        common_filter = df_2018
    elif (category == ALL) & (country == ALL):
        common_filter = df_2018[df_2018.main_category == main_category]
    elif (main_category == ALL) & (category == ALL):
        common_filter = df_2018[df_2018.country == country]
    elif (main_category == ALL) & (country == ALL):
        common_filter = df_2018[df_2018.category == category]
    elif (main_category == ALL):
        common_filter = df_2018[(df_2018.category == category) & (df_2018.country == country)]
    elif (category == ALL):
        common_filter = df_2018[(df_2018.main_category == main_category) & (df_2018.country == country)]
    elif (country == ALL):
        common_filter = df_2018[(df_2018.main_category == main_category) & (df_2018.category == category)]
    else:
        common_filter = df_2018[(main_category == main_category) & (df_2018.category == category) & (df_2018.country == country)]

# Definicja tego co ma być wyświetlane w arkuszach:  

    with output:
        global kat_glowna
        kat_glowna = main_category
        
        common_filter_view = common_filter
        common_filter_view = common_filter_view.rename(index=str,columns={'name':'Name', 'project_duration': 'Project duration' ,'category':'Category', 'main_category':'Main Category', 'currency':'Currency', 'deadline':'Deadline', 'launched':'Launch Date', 'state':'Status' ,'backers':'No. of backers','country':'Country','usd_pledged_real':'Pledged amount in USD','usd_goal_real':'Goal in USD'})
        common_filter_view = common_filter_view[['Name','Main Category','Category','Country','Currency','Status','Launch Date', 'Deadline','No. of backers', 'Project duration', 'Pledged amount in USD','Goal in USD']]
        common_filter_view = common_filter_view.sort_values('Pledged amount in USD', ascending = False).head(10)
        display(common_filter_view)
        
    with plot_output:
        plt.figure(figsize=(15,10))
        plot_output.clear_output()
        
        if country == ALL:
            
            plot_output.clear_output()
            
            df = common_filter\
            .groupby(by=["country", "state"])\
            .agg("count")\
            .reset_index()\
            .sort_values(["name", "state"], ascending = False)
            
            if not df.empty:
                plot_output.clear_output()
                sns.barplot(x='name', y="country", hue="state", data = df, palette="Set2")
                plt.xlabel("Number od projects",  fontsize=16)
                plt.ylabel("Country",  fontsize=16)
                plt.legend(loc='lower right', title = None, fontsize=14)
                plt.rc('xtick', labelsize=12)
                plt.rc('ytick', labelsize=12)
                plt.show()
                plt.close()
                pass
            else:
                plot_output.clear_output()
                print("Brak danych dla wybranych filtrów.")
        else:
            plot_output.clear_output()
            print("Aby wyświetlić ten wykres ustaw kraj na ALL.")
            pass
        
    with plot_output_1:
        
        plt.figure(figsize=(15,8))
        plot_output_1.clear_output()
        
        if ((main_category != ALL) & (category == ALL)):
            plot_output_1.clear_output()
            df_1 = common_filter[common_filter["main_category"] == main_category]\
            .groupby(by=["category", "state"])\
            .agg("count")\
            .reset_index()\
            .sort_values([ "name", "state"], ascending = False)
            if not df_1.empty:
                
                sns.barplot(x='name', y="category", hue="state", data = df_1, palette="Set2")
                plt.xlabel("Number od projects",  fontsize=16)
                plt.ylabel("Category",  fontsize=16)
                plt.legend(loc='lower right', title = None, fontsize=14)
                plt.rc('xtick', labelsize=12)
                plt.rc('ytick', labelsize=12)
                plt.show()
                plt.close()
                pass
            else:
                plot_output_1.clear_output()
                print("Brak danych dla wybranych filtrów.")

        else:
            plot_output_1.clear_output()
            print("Aby wyświetliść ten wykres wybierz kategorię główną i ustaw podkategorię na ALL. Dodatkowo możesz wybrać kraj.")
            pass
    
    with histogram_days:
        histogram_days.clear_output()
        
        if not common_filter.empty:
            histogram_days.clear_output()
            plt.figure(figsize=(15,4))
            sns.histplot(x=common_filter[common_filter["state"] == 'successful']["project_duration"],  binwidth=10, color='#FC8D62')
            plt.xlabel("Project duration (in days)", fontsize=14)
            plt.ylabel("Number of projects", fontsize=14)
            plt.title("Successful projects", fontsize=16)
            plt.show()
            
            plt.figure(figsize=(15,4))
            sns.histplot( x=common_filter[common_filter["state"] == 'failed']["project_duration"],  binwidth=10, color ='#66C2A4')
            plt.xlabel("Project duration (in days)", fontsize=14)
            plt.ylabel("Number of projects", fontsize=14)
            plt.title("Failed projects", fontsize=16)
            plt.show()
            plt.close()
        else:
            histogram_days.clear_output()
            print("Brak danych dla wybranych filtrów.")
            pass
            
    
    with data_output:
        
        data_output.clear_output()
        
        successful = len(common_filter[common_filter["state"] == 'successful'])
        failed = len(common_filter[common_filter["state"] == 'failed'])
        
        if successful != 0 and failed != 0:
            probability = str(round(int(successful)/(int(successful) + int(failed)),3)*100)+"%"
            print("Liczba podobnych zbiórek, które zakończyły się sukcesem: {},\n \nLiczba podobnych zbiórek, które zakończyły się porażką: {},\n \nProcent zbiórek podobnych zakończonych sukcesem: {}".format(successful, failed, probability))
            
            labels = 'failed', 'successful'
            x = common_filter[(common_filter["state"] == 'successful') | (common_filter["state"] == 'failed')]["state"].value_counts()
            colors = ['#66C2A4', '#FC8D62']
            explode = (0.1, 0)  # explode 1st slice
            plt.pie(x, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
            plt.axis('equal')
            plt.show()
            plt.close()
            
        elif successful == 0 and failed !=0:
            print("\n \n Liczba podobnych zbiórek wyniosła: {}. Wszystkie zbiórki zakończyły się porażką.".format(failed))
            
        elif failed == 0 and successful != 0:
            print("\n \n Liczba podobnych zbiórek wyniosła: {}. Wszystkie zbiórki zakończyły się sukcesem.".format(successful))
            
        else:
            print("Brak zbiórek dla danej selekcji. Proszę zmienić filtry.")
        
def dropdown_main_category_eventhandler(change):
    common_filtering(change.new, dropdown_category.value, dropdown_country.value )
    dropdown_category.options = ALL_values(df_2018[df_2018["main_category"] == kat_glowna].category)

def dropdown_category_eventhandler(change):
    common_filtering(dropdown_main_category.value, change.new, dropdown_country.value )

def dropdown_country_eventhandler(change):
    common_filtering(dropdown_main_category.value, dropdown_category.value, change.new)
    
    
dropdown_main_category.observe(dropdown_main_category_eventhandler, names='value')

dropdown_category.observe(dropdown_category_eventhandler, names = 'value')

dropdown_country.observe(dropdown_country_eventhandler, names='value')

# Wyświetlanie arkuszy:

item_layout = widgets.Layout(margin='0 50px 50px 0')

input_widgets = widgets.HBox([dropdown_main_category, dropdown_category, dropdown_country], layout=item_layout)

tab = widgets.Tab([output, plot_output, plot_output_1, histogram_days,  data_output], layout=item_layout)
tab.set_title(0, 'Top 10 projects')
tab.set_title(1, 'Country Plot')
tab.set_title(2, 'Category Plot')
tab.set_title(3, 'Project Duration')
tab.set_title(4, 'Stats')

dashboard = widgets.VBox([input_widgets, tab])

display(dashboard)
```
